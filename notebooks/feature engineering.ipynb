**用sklearn来做特征工程**

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

# Preprocessing
## Standardization
一般采用StandardScaler，但如果定量特征列中存在数值极大或极小的outlier，则需要用RobustScaler 。MinMaxScaler则缩放到01区间内

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler
new1 = StandardScaler().fit_transform(iris.data)
new2 = RobustScaler().fit_transform(iris.data)
new3 = MinMaxScaler().fit_transform(iris.data)
new1
```

## Normalization
标准化的目的在于统一量纲，相对于列而言，而正则化是相对于行而言，使样本在点乘、计算相似度上具有更好的特性。

```python
from sklearn.preprocessing import Normalizer
Normalizer().fit_transform(iris.data)
```

```python
Normalizer().fit_transform(new1)
```

##  Discretization
将数值型特征离散化

```python
from sklearn.preprocessing import Binarizer
Binarizer(threshold=3).fit_transform(iris.data)
```

## Encode

```python
from sklearn.preprocessing import LabelBinarizer
LabelBinarizer().fit_transform(iris.target)
```

```python
from sklearn.preprocessing import OneHotEncoder
m = OneHotEncoder().fit_transform(iris.target.reshape((-1,1)))
```

## Impute
缺失值填充

```python
from numpy import vstack, array, nan
from sklearn.preprocessing import Imputer
newdata = vstack((array([[nan, nan, nan, nan],[nan, nan, nan, nan]]), iris.data))
print(newdata[:4,:])
Imputer(missing_values="NaN",strategy="mean").fit_transform(newdata)[:4,:]
```

## Transform
对数据进行指定的变换

```python
from numpy import log1p
from sklearn.preprocessing import FunctionTransformer
#定义对数变换
FunctionTransformer(log1p).fit_transform(iris.data)
```

# Feature Select
通常来说，从两个方面考虑来选择特征：

+ 特征是否发散：如果一个特征不发散，例如方差接近于0，也就是说样本在这个特征上基本上没有差异，这个特征对于样本的区分并没有什么用。
+ 特征与目标的相关性：这点比较显见，与目标相关性高的特征，应当优选选择。一般均从相关性考虑。

根据特征选择的形式又可以将特征选择方法分为3种,我们使用sklearn中的feature_selection库来进行特征选择：

* Filter：过滤法，按照发散性或者相关性对各个特征进行评分，设定阈值或者待选择阈值的个数，选择特征。
* Wrapper：包装法，根据目标函数（通常是预测效果评分），每次选择若干特征，或者排除若干特征。
* Embedded：嵌入法，先使用某些机器学习的算法和模型进行训练，得到各个特征的权值系数，根据系数从大到小选择特征。类似于Filter方法，但是是通过训练来确定特征的优劣。
##  Filter

```python
from sklearn.feature_selection import VarianceThreshold
#参数threshold为方差的阈值
print(iris.data[:10,:])
VarianceThreshold(threshold=3).fit_transform(iris.data)[:10,:]
```
